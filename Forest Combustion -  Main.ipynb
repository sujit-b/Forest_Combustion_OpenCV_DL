```python
#importing the libraries
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Convolution2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
import warnings
warnings.filterwarnings('ignore')
```

```python
model =Sequential()
```

```python
model.add(Convolution2D(32,(3,3),input_shape=(128,128,3),activation='relu'))
```

```python
model.add(MaxPooling2D(pool_size=(2,2)))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(output_dim=150,init='uniform',activation='relu'))
```

```python
model.add(Dense(output_dim=64,init='uniform',activation='relu'))
```

```python
model.add(Dense(output_dim=1,activation='sigmoid',init='uniform'))
```

```python
from keras.preprocessing.image import ImageDataGenerator
train_datagen=ImageDataGenerator(rescale=1./255,shear_range=0.2,zoom_range=0.2,horizontal_flip=True)
test_datagen=ImageDataGenerator(rescale=1./255)
```

```python
x_train = train_datagen.flow_from_directory(r'Dataset Main\training',target_size = (128,128),batch_size = 32,class_mode = 'binary')
x_test = test_datagen.flow_from_directory(r'Dataset Main\Testing',target_size = (128,128),batch_size = 32,class_mode = 'binary')
```

```python
print(x_train.class_indices)
```

```python
model.compile(loss = 'binary_crossentropy',optimizer = "adam",metrics = ["accuracy"])
```

```python
model.fit_generator(x_train,steps_per_epoch=14,epochs=10,validation_data=x_test,validation_steps=4)
```

```python
model.save("forest.h5")
```
